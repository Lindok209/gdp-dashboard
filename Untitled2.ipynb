```python
# %% [markdown]
# ## Data Preprocessing
#
# ### Subtask:
# Prepare the data for modeling by handling categorical variables and splitting the dataset.
#
# %% [markdown]
# **Reasoning**:
# To prepare the data for machine learning models, we need to convert categorical features into numerical representations and split the dataset into training and testing sets to evaluate model performance.
#
# %%
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

# Define features (X) and target (y)
X = df.drop('default.payment.next.month', axis=1)
y = df['default.payment.next.month']

# Identify categorical and numerical features
categorical_features = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
numerical_features = [col for col in X.columns if col not in categorical_features and col != 'ID'] # Exclude 'ID' as it's not a feature

# Create a column transformer for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough' # Keep other columns (like 'ID')
)

# Create a preprocessing pipeline
preprocess_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Apply preprocessing to the features
X_processed = preprocess_pipeline.fit_transform(X)

# Get the feature names after preprocessing (optional, but helpful)
# This part can be a bit tricky with ColumnTransformer, especially with 'passthrough'
# A common approach is to use get_feature_names_out() if your scikit-learn version supports it
# or manually construct them. For simplicity here, we'll proceed without explicit names
# but keep in mind that the order of columns in X_processed corresponds to the order
# in the ColumnTransformer: numerical features first, then one-hot encoded features.

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42, stratify=y)

print("Shape of original features:", X.shape)
print("Shape of preprocessed features:", X_processed.shape)
print("Shape of training features:", X_train.shape)
print("Shape of testing features:", X_test.shape)
print("Shape of training target:", y_train.shape)
print("Shape of testing target:", y_test.shape)
```

```python
# %% [markdown]
# ## Model Building
#
# ### Subtask:
# Train a classification model and evaluate its performance.
#
# %% [markdown]
# **Reasoning**:
# Build a logistic regression model as a baseline for predicting credit card default and evaluate its performance using appropriate metrics.
#
# %%
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Initialize and train the Logistic Regression model
# Using a higher max_iter for convergence, especially with scaled data
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
print("Model: Logistic Regression")
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# %% [markdown]
# ## Model Deployment (Saving and Loading)
#
# ### Subtask:
# Save the trained model for later use and demonstrate how to load it.
#
# %% [markdown]
# **Reasoning**:
# To deploy the model or use it in a different session, we need to save its trained state. We can then load the saved model to make predictions on new data without retraining.
#
# %%
import joblib
import pickle

# Choose a filename for your model
model_filename_joblib = 'logistic_regression_model.joblib'
model_filename_pickle = 'logistic_regression_model.pkl'

# Save the model using joblib (recommended for scikit-learn models)
try:
    joblib.dump(model, model_filename_joblib)
    print(f"Model saved successfully to {model_filename_joblib}")
except Exception as e:
    print(f"Error saving model with joblib: {e}")

# You can also save using pickle, but joblib is generally preferred for scikit-learn objects
try:
    with open(model_filename_pickle, 'wb') as f:
        pickle.dump(model, f)
    print(f"Model saved successfully to {model_filename_pickle}")
except Exception as e:
    print(f"Error saving model with pickle: {e}")

# %% [markdown]
# ### Subtask:
# Load the saved model and make predictions on new data.
#
# %% [markdown]
# **Reasoning**:
# To simulate deployment, we load the saved model and demonstrate how to use it to make predictions on a new data instance.
#
# %%
# Load the model using joblib
loaded_model_joblib = joblib.load(model_filename_joblib)
print(f"\nModel loaded successfully from {model_filename_joblib}")

# Load the model using pickle
# with open(model_filename_pickle, 'rb') as f:
#     loaded_model_pickle = pickle.load(f)
# print(f"Model loaded successfully from {model_filename_pickle}")


# --- Example of making predictions with the loaded model ---

# You would typically get new data in a similar format as your original features (X)
# For demonstration, let's take the first instance from your test set (X_test)
# In a real deployment, you would process new data using the SAME preprocessing pipeline
# that you used for training data.

# Assume X_test is already preprocessed
new_data_instance = X_test[0].reshape(1, -1) # Reshape to 2D array as predict expects it

# Make a prediction
prediction = loaded_model_joblib.predict(new_data_instance)
prediction_proba = loaded_model_joblib.predict_proba(new_data_instance)

print("\nExample Prediction on a new data instance:")
print("New data instance shape:", new_data_instance.shape)
print("Predicted class:", prediction[0])
print("Prediction probabilities (Not Default, Default):", prediction_proba[0])

# To make predictions on new raw data (not yet preprocessed),
# you would need to load and apply your preprocessing pipeline as well.
# Example (assuming you saved the pipeline):
# loaded_pipeline = joblib.load('preprocessing_pipeline.joblib') # You would need to save this first
# new_raw_data = pd.DataFrame(...) # Your new raw data
# new_processed_data = loaded_pipeline.transform(new_raw_data)
# new_prediction = loaded_model_joblib.predict(new_processed_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Numerical features
numerical_features = ['LIMIT_BAL', 'AGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
for col in numerical_features:
    print(f"Descriptive Statistics for {col}:\n{df[col].describe()}\n")

# Categorical features
categorical_features = ['SEX', 'EDUCATION', 'MARRIAGE', 'default.payment.next.month']
for col in categorical_features:
    print(f"Frequency Distribution for {col}:\n{df[col].value_counts()}\n")

# Correlation matrix and heatmap
correlation_matrix = df[numerical_features].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Correlation with target variable
target_correlation = correlation_matrix['PAY_AMT6'].sort_values(ascending=False)
print("\nCorrelation with target variable 'PAY_AMT6':\n", target_correlation)
```
